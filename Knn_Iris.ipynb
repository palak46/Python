```python
import numpy as np
from numpy import genfromtxt
data=genfromtxt('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',delimiter='',dtype='unicode')

m_list=[]
for i in range(150):
    m_list.append(0)
for i in range(0,150):
    m_list[i]=str(data[i]).split(',')

my_array=np.array(m_list)

for i in range(150):
    if(my_array[i][4]=="Iris-setosa"):
        my_array[i][4]=0
    elif(my_array[i][4]=="Iris-versicolor"):
        my_array[i][4]=1
    else:
        my_array[i][4]=2
        
splittedarray=my_array[:,0:4]
colorarray=my_array[:,4]

x_train0=splittedarray[:30,:4]
x_train1=splittedarray[50:80,:4]
x_train2=splittedarray[100:130,:4]
x_train=np.concatenate((x_train0,x_train1,x_train2),axis=0)
x_train=x_train.astype(float)
x_test0=splittedarray[30:50,:4]
x_test1=splittedarray[80:100,:4]
x_test2=splittedarray[130:150,:4]
x_test=np.concatenate((x_test0,x_test1,x_test2),axis=0)
x_test=x_test.astype(float)

y_train0=colorarray[:30]
y_train1=colorarray[50:80]
y_train2=colorarray[100:130]
y_train=np.concatenate((y_train0,y_train1,y_train2),axis=0)
y_train=y_train.astype(float)
y_test0=colorarray[30:50]
y_test1=colorarray[80:100]
y_test2=colorarray[130:150]
y_test=np.concatenate((y_test0,y_test1,y_test2),axis=0)
y_train=y_train.astype(float)

tx0=x_train[:,0]
tx1=x_train[:,1]
tx2=x_train[:,2]
tx3=x_train[:,3]

def cal(x0,x1,x2,x3):
    number_of_neighbour=7
    result=np.sqrt(np.square(tx0-x0)+np.square(tx1-x1)+np.square(tx2-x2)+np.square(tx3-x3))
    sorted_result=np.sort(result)
    sorted_result=sorted_result[:number_of_neighbour]
    m_index_list=[]
    for i in range(0,number_of_neighbour):
        m_index_list.append(list(result).index(sorted_result[i]))
    values=[]
    for i in range(0,number_of_neighbour):
        values.append(y_train[m_index_list[i]])
    santosa=0
    virsicolor=0
    virginica=0
    for i in range(0,number_of_neighbour):
        if values[i]==0:
            santosa=santosa+1
        elif values[i]==1:
            virsicolor=virsicolor+1
        else:
            virginica=virginica+1
    if(santosa>(virsicolor and virginica)):
        return 0
    if(virsicolor>(santosa and virginica)):
        return 1
    if(virginica>(santosa and virsicolor)):
        return 2

m_train_list=[]
for i in range(0,60):
    x0=x_test[i,0]
    x1=x_test[i,1]
    x2=x_test[i,2]
    x3=x_test[i,3]
    m_train_list.append(cal(x0,x1,x2,x3))
    
count=0
for i in range(0,60):
    if (int(m_train_list[i])==int(y_test[i])):
        count=count+1

Accuracy=(count/len(y_test))*100
print(Accuracy)

```
